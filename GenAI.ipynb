```python
with open("dataset/The Indian Penal Code.md", 'r', encoding = 'utf-8') as f1:
    text = f1.read()
```

```python
ls = list(text)
print(ls[:100])
print("length of original list: ", len(ls))
num_of_removed = 0
```

```python
for i in ls:
    if i == '#':
        ls.remove(i)
        num_of_removed += 1
        print("removed ", i, "\n")

        
    
```

```python
print(num_of_removed)
```

```python
len_mod_ls = len(ls)
print(len_mod_ls)
```

```python
mod_text = ''.join(ls)
print(mod_text[:100])
```

```python
with open('dataset/mod_text_1.txt', 'w', encoding = 'utf8') as f:
    f.write(mod_text)
```
